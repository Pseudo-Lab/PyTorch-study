##### <span style="color:white">[Batch Noramlization : Accelerating Deep Network Training by Reducing Internal Covariate Shift]</span>

##### <span style="color:white">Normalization (정규화) : Image data의 픽셀 정보(0 ~ 255)를 (0 ~ 1.0)사이의 값으로 가지도록 제한하는 방법</span>

> Normalized data = (Image data - mean of data) / (deviation of data)
 
> 정규화를 사용하는 이유 : 
> 1. 학습 속도를 증가시킨다
> 2. Optimization 과정에서 Local minima에 빠지는 가능성을 줄여준다

##### <span style="color:white"> Internal Covariance Shift : 이전 layer의 parameter가 학습 과정에서 변함에 따라 각 layer의 input값의 분포가 변하는 현상</span>
    -> 학습 속도를 저하시키고 saturating nonlinearity를 발생시킨다
 
##### <span style="color:white"> Whitening : 각 레이어의 입력의 분산을 평균 0, 표준편차 1인 입력값으로 정규화하는 방법</span>
##### <span style="color:white"> ( Intranal Covariance Shift가 일어나지 않도록 Whitening 기법을 사용 )</span>

> Whitening의 문제점 :
> 1. covariance & inverse 계산으로 인한 계산량이 증가한다
> 2. loss function을 변화시키지 않으므로 특정 변수가 무한히 커질 가능성이 있다
> 3. 미분 불가능한 구간이 생긴다

##### <span style="color:white"> Batch Normalization (배치 정규화) : mini batch의 평균과 분산을 이용하여 정규화하여, scale, shift를 감마(γ)와 베타(β) 값으로 시키는 기법 (이때, 감마와 베타는 Backpropatation을 통해 학습)</span>

> Batch Normalization의 장점 :
> 1. 학습 속도를 증가한다
> 2. parameter initialization을 신경쓸 필요가 없다
> 3. regularization의 효과를 기대할 수 있어 dropout의 필요성이 줄어든다
> 4. gradient vanishing problem을 완화시킨다
